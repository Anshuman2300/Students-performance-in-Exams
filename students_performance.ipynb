Conclusions:-

4->Students performance depends upon parental level of education also. Those students with parental level of education of master's degree perform better than the others. those whose parental level of education is high school does not perform well.

5->those who have completed the test preparation course scores higher average score.

```python
df1=df[df['test preparation course']=='none']
```

```python
list=df['parental level of education'].unique()
average_lis=[]
for i in list:
    y=(len(df1[df1['parental level of education']==i])/len(df[df['parental level of education']==i]))*100
    average_lis.append(y)
plt.xlabel('percentage of test preparation course none')
plt.ylabel('parental level of education')
sns.barplot(y=list,x=average_lis)
plt.show()
```

-> Those with high school as  parental level has highest percentage of test preparation course none. means they did not prepared for the test.

```python
list1=df['parental level of education'].unique()
average_len=[]
for i in list1:
    y=df1[df1['parental level of education']==i]['average score'].mean()
    average_len.append(y)
plt.xlabel('average mean score of test preparation none')
plt.ylabel('parental level of education')
sns.barplot(x=average_len,y=list1)
```

-> Average mean score of those students whose parental level of education is high school is lowest among others as most of them did not prepared for the test.

->  Most students whose parental level of education is master's degree said that they did not prepared for the test yet their average mean score is highest.

-> Average mean score of those students whose parental level of education is some high school is very low even if they had prepared for the test.

```python
list2=df['race/ethnicity'].unique()
average_len=[]
for i in list2:
    y=(len(df1[df1['race/ethnicity']==i])/len(df[df['race/ethnicity']==i]))*100
    average_len.append(y)
plt.xlabel('percentage of test preparation none')
plt.ylabel('race/ethnicity')
sns.barplot(x=average_len,y=list2)
plt.show()
```
